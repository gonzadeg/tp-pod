Del archivo descargado, calcular la cantidad de veces que se dice:
*   'Quijote'
*   'mancha'
*   'españa'

Considerar que, por ejemplo, 'QuiJoTe' es lo mismo que 'quijote' (no *case sensitive*).

```python
archivo = open("pg2000.txt", "r")

contQuijote = 0
contMancha = 0
contEspaña = 0

for lineas in archivo:
  pal = lineas.split()
  for palabra in pal:
    min = palabra.lower()
    if "quijote" in min:
      contQuijote = contQuijote + 1
    elif "mancha" in min:
      contMancha = contMancha + 1
    elif "españa" in min:
      contEspaña = contEspaña + 1


print("La palabra Quijote aparece {} veces, la palabra Mancha aparece {} veces y la palabra españa aparece {} veces".format(contQuijote, contMancha, contEspaña))
```

# Base de datos - Relacionales

SQLite es una biblioteca de C que provee una base de datos ligera basada en disco que no requiere un proceso de servidor separado y permite acceder a la base de datos usando una variación no estándar del lenguaje de consulta SQL. Algunas aplicaciones pueden usar SQLite para almacenamiento interno. También es posible prototipar una aplicación usando SQLite y luego transferir el código a una base de datos más grande como PostgreSQL u Oracle.

Realizamos el `import` e indicamos el archivo en el que vamos a guardar la base de datos:

```python
import sqlite3
conn = sqlite3.connect('db.db')
```

La siguiente linea de codigo agrega filas a la misma. Por favor, ejecutar la siguiente celda para la consigna:



```python
!wget https://people.sc.fsu.edu/~jburkardt/data/csv/snakes_count_10000.csv

cur = conn.cursor()
cur.execute("CREATE TABLE t (GameNumber, GameLength);")

with open('snakes_count_10000.csv','r') as csv_file:
  lines = csv_file.readlines()[1:]
  lines = [l.split(',') for l in lines]
  to_db = [(l[0].strip(), l[1].strip()) for l in lines]

cur.executemany("INSERT INTO t (GameNumber, GameLength) VALUES (?, ?);", to_db)
conn.commit()
conn.close()
```

## Consignas

¿Como luce la base de datos? Imprimirla.


```python
conn = sqlite3.connect("db.db")
cur = conn.cursor()

cur.execute("SELECT * FROM t")
filas = cur.fetchall()

for fila in filas:
    print(fila)

conn.close()
```

¿Cuantas filas hay en la base de datos?

```python
conn = sqlite3.connect("db.db")
cur = conn.cursor()

cur.execute("SELECT COUNT(*) FROM t")
contador_fila = cur.fetchone()[0]

print("Hay {} filas en la base de datos.".format(contador_fila))

conn.close()
```

¿Cual es el máximo `GameLength` y en qué `GameNumber` se realizó?

```python
conn = sqlite3.connect("db.db")
cur = conn.cursor()

cur.execute("SELECT GameNumber, MAX(GameLength) FROM t")
max_gl = cur.fetchone()

print("El máximo GameLength es {} y se realizó en el GameNumber {}.".format(max_gl[1], max_gl[0]))
conn.close()
```

## Consigna

Hacer una request GET e imprimir uno de los campos de la respuesta:

```python
import requests

url = "https://valorant-api.com/v1/agents"

api = requests.get(url)
if api.status_code == 200:
    data = api.json()
    primer_agente = data["data"][0]["displayName"]

    print("El primer agente es: {}".format(primer_agente))
```

Hacer una request GET y, apartir de los campos en la respuesta, generar otro
 campo (por ejemplo, si la respuesta solo incluye un timestamp y una posición del registro, elaborar la velocidad del registro)

```python
import requests

url = "https://valorant-api.com/v1/agents"

response = requests.get(url)

if response.status_code == 200:
    data = response.json()
    for agente in data["data"]:
        longitudN = len(agente["displayName"])
        if longitudN > 6:
            agente["nameType"] = "Largo"
        else:
            agente["nameType"] = "Corto"
        print("Agente: {}, Tipo de nombre: {}".format(agente["displayName"], agente["nameType"]))
```

Ahora vamos a hacer algo muy parecido a lo anterior, pero vamos a usar APIs que nos ofrecen datos en tiempo real. Pueden encontrar APIs de este estilo [en este repositorio](https://github.com/ColinEberhardt/awesome-public-streaming-datasets) (pueden elegir otras que no son de ese repositorio)

## Consigna

Hacer un plot con `matplotlib` de una variable de una API de datos streaming en
el que el eje 'y' sea la variable y el eje 'x' el tiempo.

```python
import requests
import matplotlib.pyplot as plt
from datetime import datetime
import matplotlib.dates as mdates
import time

tiempo, precios = [], []

plt.ion()
fig, ax = plt.subplots()
linea, = ax.plot(tiempo, precios, marker="o")
ax.set_xlabel("Hora")
ax.set_ylabel("Precio de Bitcoin (usd)")
ax.set_title("Precio de Bitcoin (real time)")

def actualizar():
    linea.set_xdata(tiempo)
    linea.set_ydata(precios)
    ax.relim()
    ax.autoscale()
    ax.xaxis.set_major_formatter(mdates.DateFormatter("%H:%M:%S"))
    fig.autofmt_xdate()
    plt.draw()
    plt.pause(0.1)

try:
    while True:
        api = requests.get("https://api.coincap.io/v2/assets/bitcoin")
        data = api.json()
        precio = float(data["data"]["priceUsd"])
        now = datetime.now()
        tiempo.append(now)
        precios.append(precio)
        if len(tiempo) > 100:
            tiempo.pop(0)
            precios.pop(0)
        actualizar()
        time.sleep(5)
except KeyboardInterrupt:
    print("x")
```
