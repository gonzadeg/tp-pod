# Yahoo! Finance

Vamos a analizar acciones. La siguiente linea accede a Yahoo Finance y devuelve un DataFrame con los valores de la acción cada dia desde el 1980.

```python
import yfinance as yf
dataframe = yf.download('AAPL', start="1980-01-01", end="2030-01-01")
```

¿Cual es el registro mas viejo? Imprimirlo.

```python
registro_mv = dataframe.head(1)
registro_mv
```

Cree la columna `Average` tal que

$$Average =  \frac{High-Low}{2}$$

y muestre con un histograma dicha columna.

```python
dataframe["Average"] = (dataframe["High"] + dataframe["Low"]) / 2
dataframe["Average"].hist()
```

Con `matplotlib`, muestre como `Average` fue evolucionando *al final de cada año*.

```python
import matplotlib.pyplot as plt

average_por_año = dataframe["Average"].resample("Y").last()

average_por_año.plot(title="Evolución de Average")
plt.ylabel("Average")
plt.xlabel("Año")
plt.show()
```

Muestre con un gráfico de barras, como el volumen fue cambiando *año a año*.

```python
volumen = dataframe["Volume"].resample("Y").sum()

volumen.plot(kind="bar", title="Volumen por Año")
plt.ylabel("Volumen")
plt.xlabel("Año")
plt.show()
```

# Cancelaciones y Delays de vuelos del 2015

Creese una cuenta en Kaggle e importe los archivos del dataset del siguiente link: https://www.kaggle.com/datasets/usdot/flight-delays. Cree los dataframes `airlines`, `airports`, y `flights` apartir de esos archivos.

```python
airlines = pd.read_csv("/content/airlines.csv")
airports = pd.read_csv("/content/airports.csv")
flights = pd.read_csv("/content/flights.csv")
```

Combine (*join*) las tablas `airlines`, `airports`, y `flights` en una sola tabla.

```python
flights_airlines = pd.merge(flights, airlines, how= "left", left_on= "AIRLINE", right_on= "IATA_CODE")
tabla = pd.merge(flights_airlines, airports, how= "left", left_on= "ORIGIN_AIRPORT", right_on= "IATA_CODE")
tabla = pd.merge(tabla, airports, how= "left", left_on= "DESTINATION_AIRPORT", right_on= "IATA_CODE")
```

¿Cuantos vuelos fueron al aeropuerto JFK?

```python
vuelos_JFK = tabla[tabla["DESTINATION_AIRPORT"] == "JFK"].shape[0]
print("El número de vuelos que llegaron al aeropuerto JFK es: {}".format(vuelos_JFK))
```

¿Cuantos vuelos hizo la aerolinea AA?

```python
vuelos_AA = tabla[tabla["AIRLINE_x"] == "AA"]
print("Cantidad de vuelos realizados por la aerolínea AA: {}".format(vuelos_AA.shape[0]))
```

¿Que aerolineas (las primeras 5) tuvo la menor cantidad de vuelos con atrasos? Imprimirlas.

```python
atrasos = tabla[tabla["AIRLINE_DELAY"] > 0]
atrasos_airl = atrasos.groupby("AIRLINE_x")["AIRLINE_DELAY"].count().reset_index()
atrasos_airl.rename(columns={"AIRLINE_DELAY": "CANTIDAD_ATRASOS"}, inplace=True)
atrasos_airl_menor = atrasos_airl.sort_values(by="CANTIDAD_ATRASOS", ascending=True).head(5)

atrasos_airl_menor
```

¿Que aerolineas (las primeras 5) tuvo la mayor cantidad de vuelos con atrasos? Imprimirlas.

```python
atrasos_airl_mayor = atrasos_airl.sort_values(by="CANTIDAD_ATRASOS", ascending=False).head(5)
atrasos_airl_mayor
```

Haga un resumen de las razones por la cual los vuelos se atrasan.

```python
razones_atraso = [col for col in tabla.columns if "DELAY" in col.upper()]

print("Las razones por las cuales los vuelos se atrasan son:")
for razon in razones_atraso:
    print("{}".format(razon))
```

Compruebe si hay columnas con celdas vacias.

Haga una imputación de datos COMPLETA del dataframe. Pueden escojer cualquier estrategia y no necesariamente todas las columnas deben seguir la misma estrategia.

```python
numericas = tabla.select_dtypes(include="number").columns
tabla[numericas] = tabla[numericas].fillna(tabla[numericas].median())

categoricas = tabla.select_dtypes(include="object").columns
tabla[categoricas] = tabla[categoricas].fillna(tabla[categoricas].mode().iloc[0])
```
